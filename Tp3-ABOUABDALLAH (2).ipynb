## Introduction :

* En mathématiques, les équations de prédation de Lotka-Volterra, que l'on désigne aussi sous le terme de « modèle proie-prédateur », sont un couple d'équations différentielles non-linéaires du premier ordre. 

* L'objectif de ce T.P. est de mettre en oeuvre la résolution numérique d'un probléme de Cauchy avec condition initiale $y(t_0) = y_0$ pour une équation différentielle :
   $$ y_0'(t) = f(t, y(t)) $$

* Soit u et v, représentant les populations respectives de proies et de predateur satisfaisant le systéme differentiel de Lork-voltera (equation (3)) avec $k_1$ qui représente le taux de croissance des proies, $k_2$ et le taux de mortalité des prédateurs. Les proies sont dévorée proportionnellement aux produit u(t)v(t) tandis que l'abondance de nourriture profite au prédateur de maniére symétrique.

* Pour implementer cela on commence par crée notre fonction $f(t, y(t)) $ :


```julia
function f(y)
    k = [2,10]
    c = [0.001 ,0.002 ]
    u = k[1]*y[1]-c[1]*y[1]*y[2];                                                                                            
    v = -k[2]*y[2]+c[2]*y[1]*y[2]
    return [u v]
end
```

* Ensuite on implementer les différentes itérations Schéma Heun :

```julia
function iterationHeum(y,tau)
    l1 = f(y)
    l2 = f(y+l1*tau)
    return y+tau*(l1+l2)/2
end
```

```julia
tau =0.01
pop = []
lapin = []
lynx = []
y = [5000 100]
for i=1:1000
    
    append!(lynx,iterationHeum(y,tau)[2])
    append!(lapin,iterationHeum(y,tau)[1])
    
    y = iterationHeum(y,tau)
end
```

* Ici on va utiliser le module `PyPlot`afin de tracer nos coures

```julia
using PyPlot
plot(lapin)
plot(lynx)
legend(["Profile des lapins"; "Profile des lynx"])
title("Profile des deux populations")
```

#### Analyse :
* Le schéma suivant représente l'évolution des deux populations proie et prédateur. On prut voir que ses courbes sont sinusoidales et varient entre 2000 et 11000 pour les proies tandis qu'elle varient entre 100 et 9000 pour les prédateurs. Ayant les mêmes périodes.
* Le résultat suivant confirme qu'il s'agit d'une équation de lork-voltera

#### Interprétation :
* A partir du graphique obtenu on peut constater que la population de lynx évolue plus rapidement que celle des lapins mais quand les lynx atteignent leur maximum, le manque de prois conduit à une décroissance de leur population.
